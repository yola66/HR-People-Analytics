```python
#Transform the Date column as a datetime type
recruiting['Date of Application'] = pd.to_datetime(recruiting['Date of Application'], format= "%m/%d/%y")
# Convert Candidate ID Number to string
recruiting['Candidate ID Number'] = recruiting['Candidate ID Number'].astype(int)
# Convert Years of Experience to integer
recruiting['Years of Experience'] = recruiting['Years of Experience'].astype(int)

recruiting[['Date of Application','Candidate ID Number','Years of Experience']].head(10)
```

```python
# Create a new column to store the highest degree for each applicant
recruiting['Highest Degree'] = ''
for index, row in recruiting.iterrows():
    # Filter out NA values from the row and find the highest degree
    filtered_degrees = [degree for degree in row[['Degree','Degree.1', 'Degree.2', 'Degree.3']] if pd.notna(degree)]
    if filtered_degrees: 
        highest_degree = max(filtered_degrees, key=lambda x: {'Bachelors': 1, 'Masters': 2, 'JD': 2, 'PhD': 3}.get(x, 0))
        recruiting.at[index, 'Highest Degree'] = highest_degree
# Step 3: Drop the original degree columns
#recruiting.drop(columns=['Degree.1', 'Degree.2', 'Degree.3'], inplace=True)
```

```python
recruiting['Furthest Recruiting Stage Reached'] = recruiting['Furthest Recruiting Stage Reached'].replace('In-house Interview', 'In-House Interview')
Stages_adj = recruiting['Furthest Recruiting Stage Reached'].unique()
Stages_adj
```

```python
# Define the order of stages in the recruitment process
stage_order = {
    'New Application': 1,
    'Phone Screen': 2,
    'In-House Interview': 3,
    'Offer Sent': 4,
}
# Find the maximum stage reached for each candidate across all departments
max_stage_per_candidate = recruiting.groupby('Candidate ID Number')['Furthest Recruiting Stage Reached'].transform(lambda x: max(x, key=lambda stage: stage_order[stage]))

# Filter the DataFrame to keep only the rows where the stage matches the maximum stage reached
unique_candidates = recruiting[recruiting['Furthest Recruiting Stage Reached'] == max_stage_per_candidate]

unique_candidates
```

```python
# Mark duplicate rows as true
duplicate_mask = recruiting[['Candidate ID Number','Department','Furthest Recruiting Stage Reached']].duplicated(keep=False)

# Select all rows with duplicates
duplicates = recruiting.loc[duplicate_mask]
duplicates

# Filter the DataFrame to only include rows where 'Furthest Recruiting Stage Reached' is 'Offer Sent 
offer_sent_duplicates= duplicates[duplicates['Furthest Recruiting Stage Reached'] == 'Offer Sent']
offer_sent_duplicates.shape[0]
# no people apply several position within department reach the offer stage
```
